```python
########## visualization
from graphviz import Digraph
from anytree.exporter import DotExporter

def render_tree(root):
    
    dot = Digraph()

    
    for pre, fill, node in RenderTree(root):
        dot.node(node.name)

    
    for node in root.descendants:
        for child in node.children:
            dot.edge(node.name, child.name)

    
    dot.render('tree', format='png')
```

```python
############initialize


class MyNode(NodeMixin):
    def __init__(self, name,parent, **kwargs):
        super(MyNode, self).__init__()
        self.name = name
        self.parent = parent
        self.__dict__.update(kwargs)
```

```python
########## إرجاع ال nodes 
def get_nodes_at_level(root, level):
   
    if level == 0:
        return [root]
    
    nodes = []
    for child in root.children:
        nodes.extend(get_nodes_at_level(child, level - 1))
    
    return nodes
```

```python
########### إضافة node  جديدة بدون ملف 
def insert_node_by_path(root, path, node_name):
   
    path_list = path.split('/')
    
    
    current_node = root
    for name in path_list:
        
        child_node = None
        for child in current_node.children:
            if child.name == name:
                child_node = child
                break
        
        
        if child_node is None:
            child_node = Node(name, parent=current_node)
        
       
        current_node = child_node
    
    
    MyNode(node_name, parent=current_node)



```

```python
### حذف بناء على الإسم
from anytree import Node, RenderTree

def delete_by_name(node_name):

    found_node = None
    for node in root.descendants:
        if node.name == node_name:
            found_node = node
            break

    if found_node is None:
        raise ValueError(f"Node {node_name} not found in the tree")

    for child in found_node.children:
        child.parent = None

    found_node.parent = None


```

```python

######### حذف حسب مسار ال node
def delete_by_name_path(node_path):

    found_node = root
    for name in node_path.split("/")[1:]:
        found_node = next((node for node in found_node.children if node.name == name), None)
        if found_node is None:
            raise ValueError(f"Node {name} not found in the tree")

    if not found_node.children:
        raise ValueError(f"Node {node_path} does not have any children")

    for child in found_node.children:
        child.parent = None

    found_node.parent = None
```
